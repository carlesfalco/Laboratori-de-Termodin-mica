```python
from math import pi
from numpy import array, mean, std, sqrt, log, linspace, exp, polyfit, diag, loadtxt, concatenate, load, save
from scipy.stats import linregress
import matplotlib.pyplot as plt
from matplotlib import rcParams
import matplotlib.pylab as pl
```

```python
titanium = load('titanium_mat.npy')
gold = load('gold_mat.npy')
aluminum = load('aluminum_mat.npy')
iron = load('iron_mat.npy')
copper = load('copper_mat.npy')
silver = load('silver_mat.npy')
lead = load('lead_mat.npy')

rcParams.update({'font.size': 18})
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

plt.figure(figsize = (10,5))
dspan = linspace(0,1,len(titanium))
plt.ylabel('$T(x,t = 1$ h$)\;(^\circ$C)')
plt.xlabel('Dist\`ancia (m)')

colors = pl.cm.viridis(linspace(1,0,7))

plt.plot(dspan,silver,linewidth = 1.1, label = 'Plata', color = colors[0])
plt.plot(dspan,copper,linewidth = 1.1, label = 'Coure', color = colors[1])
plt.plot(dspan,gold,linewidth = 1.1, label = 'Or', color = colors[2])
plt.plot(dspan,aluminum,linewidth = 1.1, label = 'Alumini', color = colors[3])
plt.plot(dspan,iron,linewidth = 1.1, label = 'Ferro', color = colors[4])
plt.plot(dspan,lead,linewidth = 1.1, label = 'Plom', color = colors[5])
plt.plot(dspan,titanium,linewidth = 1.1, label = 'Titani', color = colors[6])
leg = plt.legend(prop={'size': 15})

for line in leg.get_lines():
    line.set_linewidth(2.0)
    
plt.savefig('simu3.pdf',bbox_inches = 'tight')
plt.show()
```

```python
titanium = load('ti_pos.npy')
gold = load('au_pos.npy')
aluminum = load('al_pos.npy')
iron = load('iron_pos.npy')
copper = load('cu_pos.npy')
silver = load('silver_pos.npy')
lead = load('pb_pos.npy')
tspan = linspace(0,9000-1709,len(lead))

vect = [titanium, gold, aluminum, iron, copper, silver, lead]
labels = ['Titani','Or','Alumini','Ferro','Cobre','Plata','Plom']
z = list(zip(vect,labels))
z = sorted(z, key = lambda x: min(x[0]))
z = array(z)
labels = z[:,1]
vect = z[:,0]

rcParams.update({'font.size': 18})
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

plt.figure(figsize = (10,5))
dspan = linspace(1,0,len(titanium))
plt.ylabel('$T(x = 8$ cm$,t)\;(^\circ$C)')
plt.xlabel('Temps (min)')

colors = pl.cm.viridis(linspace(1,0,len(vect)))

i = 0
for v in vect:
    plt.plot(tspan/60, v, linewidth = 1.1, color = colors[i], label = labels[i])
    i += 1

leg = plt.legend(prop={'size': 15}, loc = 'upper center',bbox_to_anchor = [0.5,1.23], ncol = 4)

for line in leg.get_lines():
    line.set_linewidth(2.0)

plt.savefig('simu_materials_posicio.pdf',bbox_inches = 'tight')
plt.show()
```
